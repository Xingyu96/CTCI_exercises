```python
class Node():
    def __init__(self,value,nextNode=None):
        self.value = value
        self.nextNode = nextNode
    def __str__(self):
        return str(self.value)

class Stack():
    def __init__(self,size=0,top=None):
        self.size = size
        self.top = top
    def __str__(self):
        #print whole thing
        if self.top == None:
            return ""
        elif self.top.nextNode == None:
            return str(self.top.value)
        else:
            temp = self.top
            s = ""
            while temp != None:
                s += (str(temp) + " ")
                temp = temp.nextNode
            return s
    def __len__(self):
        return self.size
    
    def pop(self):
        if self.size == 0:
            return None
        else:
            temp = self.top
            self.top = self.top.nextNode
            self.size -= 1
            return temp
        
    def push(self,value):
        n = Node(value)
        if self.size == 0:
            self.top = n
            self.size += 1
        else:
            n.nextNode = self.top
            self.top = n
            self.size += 1
        
    def peek(self):
        return self.top.value
        
    def isEmpty(self):
        return self.size == 0
                    
class Queue():
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    def __str__(self):
        temp_node = self.head
        s = ""
        for i in range(self.size):
            s += (str(temp_node) + " ")
            temp_node = temp_node.nextNode
        return s
    def __len__(self):
        return self.size
    
    def enqueue(self, value):
        n = Node(value)
        if self.size == 0:
            self.head = n
            self.tail = n
        else:
            self.tail.nextNode = n 
            self.tail = n
        self.size += 1
    def dequeue(self):
        if self.size == 0:
            return None
        elif self.size == 1:
            n = self.head
            self.head = None
            self.tail = None
            self.size -= 1
            return n
        else:
            n = self.head
            self.head = n.nextNode
            self.size -= 1
            return n
    def peek(self):
        return self.head.value
    def isEmpty(self):
        return self.size == 0

#3.2 Min Stack
class MinStack(Stack):
    def __init__(self):
        self.size = 0
        self.top = None
        self.__minStack = Stack()
    def pop(self):
        if self.size == 0:
            return None
        else:
            temp = self.top
            self.top = self.top.nextNode
            self.size -= 1
            if temp.value == self.__minStack.peek():
                self.__minStack.pop()
            return temp
    def push(self, value):
        n = Node(value)
        if self.size == 0:
            self.top = n
            self.size += 1
            self.__minStack.push(n.value)
        else:
            n.nextNode = self.top
            self.top = n
            self.size += 1
            if n.value < self.__minStack.peek():
                self.__minStack.push(n.value)
    def minimum(self):
        return self.__minStack.peek()

#3.3 Stack-o-Stacks
class Stack_of_Stacks(Stack):
    def __init__(self,stackSize,numStacks):
        self.stackArray = []
        for i in range(numStacks):
            self.stack_array[i] = Stack()
        self.stackSize = stackSize
        self.size = 0
        self.top = None
        
    def pop(self):
        return False
        
    def push(self,value):
        return False
        
    def popAt(self,stackNum):
        return False

#3.4 Stack Queue
class StackQueue(Stack):
    #2 stacks growing in opposite directions
    #enqueue from head_stack
    #dequeue from tail_stack
    #if tail_stack empty on dequeue, pop everything from head_stack into tail_stack
    def __init__(self):
        self.head_stack = Stack()
        self.tail_stack = Stack()
        self.size = 0
    def __str__(self):
        reverse_head = Stack()
        for i in range(self.head_stack.size):
            reverse_head.push(self.head_stack.pop().value)
        s = ""
        temp = self.tail_stack.top
        for i in range(self.tail_stack.size):
            s += (str(temp.value) + " ")
        temp = reverse_head.top
        for j in range(reverse_head.size):
            s += (str(temp.value) + " ")
        return s
        
    def enqueue(self,value):
        self.head_stack.push(value);
        self.size += 1
    def dequeue(self):
        if self.tail_stack.size == 0:
            for i in range(self.head_stack.size):
                self.tail_stack.push(self.head_stack)
            self.size -= 1
            return self.tail_stack.pop()
        else:
            self.size -= 1
            return self.tail_stack.pop()
        
    def peek(self):
        if self.tail_stack.size == 0:
            for i in range(self.head_stack.size):
                self.tail_stack.push(self.head_stack)
            return self.tail_stack.top.value
        else:
            return self.tail_stack.top.value
    
        
```
