```python
class Node():
    def __init__(self,value,nextNode=None):
        self.value = value
        self.nextNode = nextNode
    def __str__(self):
        return str(self.value)
    
class SinglyLinkedList():
    def __init__(self):
        self.head = None
        self.size = 0
    def __str__(self):
        #print whole thing
        if self.head == None:
            return ""
        elif self.head.nextNode == None:
            return str(self.head.value)
        else:
            temp = self.head
            s = ""
            while temp != None:
                s += (str(temp) + " ")
                temp = temp.nextNode
            s = s[:len(s) - 1]
            return s
    def __len__(self):
        return self.size
    def append(self, value):
        n = Node(value)
        tmp = self.head
        if self.size == 0:
            self.head = n
        elif self.size == 1:
            self.head.nextNode = n
        else:
            for i in range(self.size - 1):
                tmp = tmp.nextNode
            tmp.nextNode = n

        self.size += 1
    def pop(self):
        temp = self.head
        while temp.nextNode.nextNode != None:
            temp = temp.nextNode
        n = temp.nextNode
        temp.nextNode = None
        self.size -= 1
        return n
        
    #2.2 Nth to the last
    def n_to_last(self, n):
        if n > self.size:
            return None
        elif self.size == 0:
            return None
        elif self.size == 1:
            return self.head
        else:
            temp = self.head
            for i in range(self.size - n - 1):
                temp = temp.nextNode
            return temp
    #2.3 Remove Middle
    def remove_middle_node(self):
        if self.size == 0:
            return None
        elif self.size == 1:
            self.head = None
            self.size -= 1
        else:
            middle = int(self.size / 2)
            temp = self.head
            for i in range(middle - 2):
                temp = temp.nextNode
            todelete = temp.nextNode
            temp.nextNode = todelete.nextNode
            todelete = None
            self.size -= 1
        
    #2.4
    def partition(self, cutoff):
        smaller_list = SinglyLinkedList()
        bigger_list = SinglyLinkedList()
        equal_list = SinglyLinkedList()
        smaller_end = self.head
        equal_end = self.head
        temp = self.head 
        for i in range(self.size):
            if temp.value > cutoff:
                bigger_list.append(temp.value)
            elif temp.value == cutoff:
                equal_list.append(temp.value)
                equal_end = temp
            else:
                smaller_list.append(temp.value)
                small_end = temp
            temp = temp.nextNode
        combined_list = join(smaller_list, equal_list)
        combined_list = join(combined_list, bigger_list)
        return combined_list
            
def join(list_a, list_b):
    combined_list = SinglyLinkedList()
    temp = list_a.head
    for i in range(list_a.size):
        combined_list.append(temp.value)
        temp = temp.nextNode
    temp = list_b.head
    for j in range(list_b.size):
        combined_list.append(temp.value)
        temp = temp.nextNode
    return combined_list
            
#2.5
def sum_lists(list_a, list_b):
    a = 0
    b = 0
    c = 0
    multiplier_a = 1
    multiplier_b = 1
    
    temp_a = list_a.head
    temp_b = list_b.head
    
    while temp_a != None:
        a += temp_a.value * multiplier_a
        multiplier_a *= 10
        temp_a = temp_a.nextNode
        
    while temp_b != None:
        b += temp_b.value * multiplier_b
        multiplier_b *= 10
        temp_b = temp_b.nextNode
    
    c = a + b
    
    sum_list = SinglyLinkedList()
    while c != 0:
        sum_list.append(c % 10)
        c =  int((c - c % 10)/10)
    return sum_list
    
#2.6
def is_palindrome(l):
    letter_stack = SinglyLinkedList()
    temp = l.head
    isEven = (l.size % 2 == 0)
    for i in range(int(l.size / 2)):
        letter_stack.append(temp.value)
        temp = temp.nextNode
    if isEven == False:
        temp = temp.nextNode
    for j in range(int(l.size / 2)):
        popped = letter_stack.pop()
        if popped.value != temp.value:
            return False
    return True
                
        
```
