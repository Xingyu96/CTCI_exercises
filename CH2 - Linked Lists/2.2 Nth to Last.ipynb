```python
class Node():
    def __init__(self,value,nextNode=None):
        self.value = value
        self.nextNode = nextNode
    def __str__(self):
        return str(self.value)
    
class SinglyLinkedList():
    def __init__(self):
        self.head = None
        self.size = 0
    def __str__(self):
        #print whole thing
        if self.head == None:
            return ""
        elif self.head.nextNode == None:
            return str(self.head.value)
        else:
            temp = self.head
            s = ""
            while temp != None:
                s += (str(temp) + " ")
                temp = temp.nextNode
            s = s[:len(s) - 1]
            return s
    def __len__(self):
        return self.size
    def append(self, value):
        n = Node(value)
        tmp = self.head
        if self.size == 0:
            self.head = n
        elif self.size == 1:
            self.head.nextNode = n
        else:
            for i in range(self.size - 1):
                tmp = tmp.nextNode
            tmp.nextNode = n

        self.size += 1
    def pop(self):
        if self.size == 0:
            return None
        elif self.size == 1:
            n = self.head
            self.head = None
            self.size = 0
            return n
        else:
            temp = self.head
            while temp.nextNode.nextNode != None:
                temp = temp.nextNode
            n = temp.nextNode
            temp.nextNode = None
            self.size -= 1
            return n
        
    #2.2 Nth to the last
    def n_to_last(self, n):
        if n > self.size:
            return None
        elif self.size == 0:
            return None
        elif self.size == 1:
            return self.head
        else:
            temp = self.head
            for i in range(self.size - n - 1):
                temp = temp.nextNode
            return temp
    #2.3 Remove Middle
    def remove_middle_node(self):
        if self.size == 0:
            return None
        elif self.size == 1:
            self.head = None
            self.size -= 1
        else:
            middle = int(self.size / 2)
            temp = self.head
            for i in range(middle - 2):
                temp = temp.nextNode
            todelete = temp.nextNode
            temp.nextNode = todelete.nextNode
            todelete = None
            self.size -= 1
        
    #2.4
    def partition(self, cutoff):
        smaller_list = SinglyLinkedList()
        bigger_list = SinglyLinkedList()
        equal_list = SinglyLinkedList()
        smaller_end = self.head
        equal_end = self.head
        temp = self.head 
        for i in range(self.size):
            if temp.value > cutoff:
                bigger_list.append(temp.value)
            elif temp.value == cutoff:
                equal_list.append(temp.value)
                equal_end = temp
            else:
                smaller_list.append(temp.value)
                small_end = temp
            temp = temp.nextNode
        combined_list = join(smaller_list, equal_list)
        combined_list = join(combined_list, bigger_list)
        return combined_list
            
def join(list_a, list_b):
    combined_list = SinglyLinkedList()
    temp = list_a.head
    for i in range(list_a.size):
        combined_list.append(temp.value)
        temp = temp.nextNode
    temp = list_b.head
    for j in range(list_b.size):
        combined_list.append(temp.value)
        temp = temp.nextNode
    return combined_list
            
#2.5
def sum_lists(list_a, list_b):
    a = 0
    b = 0
    c = 0
    multiplier_a = 1
    multiplier_b = 1
    
    temp_a = list_a.head
    temp_b = list_b.head
    
    while temp_a != None:
        a += temp_a.value * multiplier_a
        multiplier_a *= 10
        temp_a = temp_a.nextNode
        
    while temp_b != None:
        b += temp_b.value * multiplier_b
        multiplier_b *= 10
        temp_b = temp_b.nextNode
    
    c = a + b
    
    sum_list = SinglyLinkedList()
    while c != 0:
        sum_list.append(c % 10)
        c =  int((c - c % 10)/10)
    return sum_list
    
#2.6
def is_palindrome(l):
    if l.size == 0:
        return False
    elif l.size == 1:
        return True
    else:
        letter_stack = SinglyLinkedList()
        temp = l.head
        for i in range(int(l.size / 2)):
            letter_stack.append(temp.value)
            temp = temp.nextNode
        print(letter_stack)
        if l.size % 2 == 1:
            temp = temp.nextNode
        for j in range(int(l.size / 2)):
            popped = letter_stack.pop()
            if popped.value != temp.value:
                return False
        return True
#2.7 - remember, if they intersect they share the rest of the list
def do_they_intersect(a, b):
    #initial variables
    t_a = a.head
    t_b = b.head
    
    #compare last nodes, if not same then don't intersect
    for i in range(a.size - 1):
        t_a = t_a.nextNode
    for j in range(b.size - 1):
        t_b = t_b.nextNode
    
    if t_a is t_b == False:
        return None
    
    #reset variables
    else:
        t_a = a.head
        t_b = b.head
        diff = t_a.size - t_b.size
        if diff < 0:
            for i in range(diff,0):
                t_b = t_b.nextNode
        elif diff > 0:
            for i in range(diff,0):
                t_a = t_a.nextNode
        while t_a is t_b != True:
            t_a = t_a.nextNode()
            t_b = t_b.nextNode()
        return t_a
    
#2.8 - Detect Loops
def is_circular(l):
    slow_hop = l.head
    fast_hop = l.head.nextNode.nextNode
    
    while(fast_hop.nextNode.nextNode != None):
        if slow_hop is fast_hop:
            return True
        elif fast_hop.nextNode is slow_hop:
            return True
        else:
            slow_hop = slow_hop.nextNode
            fast_hop = fast_hop.nextNode.nextNode
    return False
        
        
```
